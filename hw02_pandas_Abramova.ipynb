средний балл за экзамен по чтению для каждой этнический группы

```python
%time
readingscores = [0, 0, 0, 0, 0]
groupscount = [0, 0, 0, 0, 0]
for i in range(df.shape[0]):
  if df.iloc[i]['race/ethnicity'] == 'group A':
    readingscores[0] += int(df.iloc[i]['reading score'])
    groupscount[0] += 1
  if df.iloc[i]['race/ethnicity'] == 'group B':
    readingscores[1] += int(df.iloc[i]['reading score'])
    groupscount[1] += 1
  if df.iloc[i]['race/ethnicity'] == 'group C':
    readingscores[2] += int(df.iloc[i]['reading score'])
    groupscount[2] += 1
  if df.iloc[i]['race/ethnicity'] == 'group D':
    readingscores[3] += int(df.iloc[i]['reading score'])
    groupscount[3] += 1
  if df.iloc[i]['race/ethnicity'] == 'group E':
    readingscores[4] += int(df.iloc[i]['reading score'])
    groupscount[4] += 1
means = [0, 0, 0, 0, 0]
for i in range (5):
  means[i] = readingscores[i]/groupscount[i]
pd.DataFrame([['group A', means[0]], ['group B', means[1]], ['group C', means[2]], ['group D', means[3]], ['group E', means[4]]], columns=['race/ethnicity', 'mean'])
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
      return 'A'
    if average_mark >= 80:
      return 'B'
    if average_mark >= 70:
      return 'C'
    if average_mark >= 60:
      return 'D'
    if average_mark >= 50:
      return 'E'
    if average_mark < 50:
      return 'F'

df["average_mark"] = df.apply(lambda row: (row["math score"] + row["reading score"] + row["writing score"])/3, axis = 1)
df["Grade"] = df['average_mark'].apply(GetGrade)
df.head(2)
```
